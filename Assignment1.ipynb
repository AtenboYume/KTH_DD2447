## 1.2 What if you are interested in a distribution which is not covered in the standard programming libraries?

Let $p(x)$ be a non-standard and unnormalized distribution, say

$p(x) = e^{-\alpha x} e^{-\lambda x}$, $x \geq 0$

then how can we sample from it? Although it is not a complex distribution, it is not available in `numpy` nor `sklearn`.

We can resolve the dilemma via a clever trick using the cumulative density function (CDF) and the uniform distribution, $\mathcal{U}([0,1])$. But first, you will need to normalize $p(x)$.

Read the instructions in Murphy (ch. 23.2.1) to sample from the normalized $p(x)$. Use $\alpha=0.5$ and $\lambda=1.0$ and visualize the samples below in the same manner as you have done above (but for a single parameter setting, obviously). Also, include your derivation in the text cell below.

**Hint:** to normalize a distribution, you need to compute the normalizing constant $$C=\int_{\mathbb{R}} p(x) dx,$$
such that
$$\int_{\mathbb{R}} p(x) / C dx = 1.$$


$p(x) = e^{-\alpha x} e^{-\lambda x}, \quad x \geq 0$


To normalize this distribution, find the normalizing constant C  such that

$C = \int_{\mathbb{R}} p(x) \, dx$

Compute the integral step by step:

$C = \int_{0}^{\infty} e^{-\alpha x} e^{-\lambda x} \, dx$

$C = \int_{0}^{\infty} e^{-(\alpha + \lambda) x} \, dx$

Now, solve the integral:

$C = \left[ -\frac{1}{\alpha + \lambda} e^{-(\alpha + \lambda) x} \right]_{0}^{\infty}$

Taking the limit as x approaches infinity:

$C = \lim_{{x \to \infty}} \left( -\frac{1}{\alpha + \lambda} e^{-(\alpha + \lambda) x} \right)$

Since  $\alpha$ and  $\lambda$ are positive, the exponential term approaches zero as x goes to infinity:

$C = \frac{1}{\alpha + \lambda}$

So, the normalizing constant C is given by:

$C = \frac{1}{\alpha + \lambda}$


````python
# code to sample from and visualize p(x) here
##parameters
alpha, lamda = 0.5, 1
C = 2/3

##number of samples
num_samples = 1000
uniform_samples = np.random.rand(num_samples)

## inversed CDF
def inversed_CDF(u):
    return -np.log(1-u)*C

##PDF of the normalized distribution
def normalized_distribution(x):
    return np.exp(-alpha * x) * np.exp(-lamda * x) / C

x = inversed_CDF(uniform_samples)
Y = normalized_distribution(x)

x_values = np.linspace(0, 5, 1000)
plt.plot(x_values,normalized_distribution(x_values),label='Normalized PDF')
plt.hist(x, bins=30, density=True, alpha=0.5, label='Generated Samples')
plt.title('Sampling from Non-standard Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
````

## 4.1 Approximate the mean of a LogNormal distribution using the Law of Large Numbers
The LogNormal is a very important distribution, which can be used to model the distribution of quantities such as time, the values of purchases, or distances.

Here we assume that we are sequentially obtaining data that we believe are coming from a LogNormal distribution. Eventually, we will receive a sufficient amount of data to accurately approximate the true mean of the LogNormal distribution. This follows from the Law of Large Numbers:

Let, $$a = \mathbb{E}_{p(x)}[g(x)]$$ and $$\hat{a}_n = \frac{1}{n}\sum_{i=1}^nx_i,$$ where $x_i\sim p(x)$, then $$\lim_{n→\infty}\hat{a}_n = a.$$ Meanwhile, for all finite (natural) numbers of $n$, we say that $$\hat{a}_n\approx a,$$ or, in words, $\hat{a}_n$ is a Monte Carlo estimate of $a$.

To sample from a LogNormal, use the following scheme.

Let $$z_i \sim \mathcal{N}(\mu, \sigma^2),$$ and set $$x_i = \exp(z_i).$$ This is equivalent to sampling $x_i$ from $\log\mathcal{N}(\mu, \sigma^2)$.

### 4.1.1

Let $\mu = 1$ and $\sigma^2 = 5$, and set $p(x|\mu, \sigma^2) = \log\mathcal{N}(\mu, \sigma^2)$. Compute the mean of the LogNormal distribution in the cell below. This should be done using the expression for the mean of the LogNormal (i.e., do not approximate it). *Hint: it's not equal to $\mu$*.

$m = \mathbb{E}_{p(x|\mu, \sigma^2)}[x]$ = $e^{\mu + \frac{\sigma^2}{2}}$ = $e^{\frac{7}{2}}$

````python
# parameters
mu = 1
sigma_2 = 5

# Calculate the mean of the LogNormal distribution
mean_lognormal = np.exp(mu + 0.5 * sigma_2)

# Print the result
print(f"The mean of the LogNormal distribution is: {mean_lognormal}")
````

### 4.1.2
Use the provided plotting function to plot how the Monte Carlo approximation improves with more data.

First you will need to sample $n=100,000$ samples from $p(x|\mu, \sigma^2)$ using numpy and the scheme provided above. Observe that when sampling from a normal distribution in numpy, the second argument is the standard deviation not the variance.

````python
import matplotlib.pyplot as plt
import numpy as np

# Function to plot sequential Monte Carlo estimates
def plot_means(m, data):
    n = len(data)
    sequential_estimates = []
    for i in range(1, n + 1):
        m_hat_i = np.mean(data[:i])  # compute the Monte Carlo approximation of m using only the data points x_1,..., x_i
        sequential_estimates.append(m_hat_i)
    plt.plot(sequential_estimates, label=r'$\hat{m}_i$')
    plt.axhline(m, 0, n, color='red', label='$m$')
    plt.xlabel('Number of Samples')
    plt.ylabel('Monte Carlo Approximation')
    plt.title('Sequential Monte Carlo Approximation Convergence')
    plt.legend(loc='best')
    plt.show()

def lognormal(mu,sigma,n_samples):
    np.random.seed(0)
    z_samples = np.random.normal(mu,sigma, size=n_samples)
    x_samples = exp(z_samples)
    return z_samples, x_samples

# Plot the Monte Carlo approximation
x_samples = lognormal(1,np.sqrt(5),100000)[1]
plot_means(mean_lognormal, x_samples)
````

### 4.1.3
Repeat the experiment above but now use $\sigma^2 = 1$. Plot the approximations, and explain why the approximations converge with less data. Reflect on the limitation of the Law of Large Numbers here.

````python
x_samples = lognormal(1,1,100000)[1]
plot_means(exp(3/2),x_samples)

x_samples = lognormal(1,1,10000)[1]
plot_means(exp(3/2),x_samples)

x_samples = lognormal(1,1,1000)[1]
plot_means(exp(3/2),x_samples)

x_samples = lognormal(1,1,100)[1]
plot_means(exp(3/2),x_samples)
````

As the sample size increases, the Monte Carlo approximations converge to the true mean. This is consistent with the Law of Large Numbers, which states that the sample mean converges to the true mean as the sample size grows.

Limitations of the Law of Large Numbers:

The Law of Large Numbers guarantees convergence as the sample size approaches infinity. However, for finite sample sizes, there can be fluctuations and variations in the estimates, especially when dealing with distributions that have heavy tails or outliers.
Impact of Outliers:

Outliers or extreme values can have a more significant impact on smaller sample sizes, leading to larger fluctuations in the estimates. This is a limitation of the Law of Large Numbers, and practical applications may need to consider the influence of outliers.

1.  Model the target pdf, $p(x)$. You can use an indicator function. Be clear about what the corresponding "indicator" set is.

2.  Draw 20k importance samples from $q(x)$ and compute the corresponding *importance weights*, using $p(x)$ as the target distribution.

3.  Plot the importance samples in a scatter plot, using the corresponding importance weights as colors for the different points. The result should be a non-uniform distribution over the disk. Why is it non-uniform? Explain the distribution of importance weights.

4. How can you change the parameters in $q(x)$ in order to make the distribution of importance weights more uniform over the disk?

Indicator: *$\mathbb{1}_{\{x^2 + y^2 <= 0.5^2\}}(x,y)$*

````python
# compute the importance weights and plot the colored points here
def target_pdf(x):
    return 1 if x[0]**2 + x[1]**2 <= 0.5**2 else 0
````

2.

````python
# Draw 20,000 samples from q(x) and compute importance weights
np.random.seed(0)  # for reproducibility
num_samples = 20000

# Proposal distribution q(x) parameters
mu_q = np.array([0, 0])
cov_q = np.array([[0.5, 0], [0, 0.5]])

# Draw samples from q(x)
samples_q = np.random.multivariate_normal(mu_q, cov_q, num_samples)

# Compute importance weights
weights = np.array([target_pdf(sample) / (0.5 * 0.5) for sample in samples_q])
````

3

````python
# Plot the importance samples
plt.scatter(samples_q[:, 0], samples_q[:, 1], c=weights, alpha=0.5)
plt.title('Importance Sampling: Non-uniform Distribution over the Disk')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(label='Importance Weights')
plt.show()
````

*Answer the questions in 3 here.*

The distribution is non-uniform because importance samples are more likely to fall in regions of the disk with higher probability density. The center of the disk has higher density, leading to more samples in that region.

````python
# Adjust the covariance matrix to make the proposal distribution more diffuse
cov_q_adjusted = np.array([[2.0, 0], [0, 2.0]])

# Draw samples from the adjusted q(x)
samples_q_adjusted = np.random.multivariate_normal(mu_q, cov_q_adjusted, num_samples)

# Compute importance weights for the adjusted distribution
weights_adjusted = np.array([target_pdf(sample) / (2.0 * 2.0) for sample in samples_q_adjusted])

# Plot the importance samples from the adjusted distribution
plt.scatter(samples_q_adjusted[:, 0], samples_q_adjusted[:, 1], c=weights_adjusted, alpha=0.5)
plt.title('Importance Sampling: Uniform Distribution over the Disk')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(label='Importance Weights')
plt.show()
````
